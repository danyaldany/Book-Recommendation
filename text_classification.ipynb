```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli",
                     device='cpu')
```

```python
fiction_categories = ['Fiction', 'NonFiction']

sequence = books.loc[books['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)[0]
# Jahan bhi category 'Fiction' hogi, us row ki description column se value lo. 
# Phir pehli wali ko sequence me daalo.
```

```python
classifier(sequence, fiction_categories)
```

```python
import numpy as np

#maximum score ko  max_index ma save karaga  
max_index = np.argmax(classifier(sequence, fiction_categories)['scores'])
max_label = classifier(sequence, fiction_categories)['labels'][max_index]
print(max_label)
print(max_index)
```

```python
#we create a function to return the max_index AND max_label
```

```python
from tqdm import tqdm
import numpy as np
import pandas as pd

fiction_categories = ["Fiction", "NonFiction"]

def generate_prediction(sequence, candidate_labels):
    prediction = classifier(
        sequence,
        candidate_labels,
        hypothesis_template="This book is {}."
    )
    max_index = np.argmax(prediction['scores'])
    return prediction['labels'][max_index]

```

```python
actual_cats = []
prediction_cats = []

#Fiction
fiction_books = books.loc[books['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)

for i in tqdm(range(min(100, len(fiction_books)))):
    sequence = fiction_books[i]
    prediction_cats.append(generate_prediction(sequence, fiction_categories))
    actual_cats.append('Fiction')

#NonFiction
nonfiction_books = books.loc[books['simple_categories'] == 'NonFiction', 'description'].reset_index(drop=True)

for i in tqdm(range(min(100, len(nonfiction_books)))):
    sequence = nonfiction_books[i]
    prediction_cats.append(generate_prediction(sequence, fiction_categories))
    actual_cats.append('NonFiction')
```

```python
# Build DataFrame
prediction_df = pd.DataFrame({
    "Actual_Categories": actual_cats,
    "Prediction_Categories": prediction_cats
})

# Check distribution
print(prediction_df.value_counts().reset_index())

# Correct prediction flag
prediction_df['correct_pred'] = np.where(
    prediction_df['Actual_Categories'] == prediction_df['Prediction_Categories'],
    1,
    0
)
```

```python
# Accuracy
accuracy = prediction_df['correct_pred'].mean()
print("Accuracy:", accuracy)
```

```python
#or 
#used to find the accuracy

prediction_df['correct_pred'].sum()/len(prediction_df)
```

```python
#this will return the missing(null) rows and columns of isbna and description
missing_cats = books.loc[books['simple_categories'].isna(), ['isbn13', 'description']].reset_index(drop=True)
```

```python
missing_cats.shape
```

```python
prediction_cats = []
isbn = []

for i in tqdm(range(len(missing_cats))):
    sequence = missing_cats['description'].iloc[i]
    prediction = generate_prediction(sequence, fiction_categories)
    
    prediction_cats.append(prediction)   # âœ… full string
    isbn.append(missing_cats['isbn13'].iloc[i])
```

```python
print(len(isbn))
print(len(prediction_cats))
```

```python
min_len = min(len(isbn), len(prediction_cats))
missing_predicted_df = pd.DataFrame({
    'isbn13': isbn[:min_len],
    'prediction_categories': prediction_cats[:min_len]
})
```

```python
# Fix split words by grouping rows per ISBN
fixed_df = (missing_predicted_df
            .groupby('isbn13')['prediction_categories']
            .apply(''.join)   # join chars back into full word
            .reset_index())
fixed_df
```

```python
books = pd.merge(books, fixed_df, on = 'isbn13',how='left')
```

```python
books['simple_categories'] = books['simple_categories'].fillna(books['prediction_categories'])
books = books.drop(columns=['prediction_categories'])
```
